```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("hourly_fitbit_sema_df_unprocessed.csv")
df
```

```python
id_counts = (
    df["id"]
    .value_counts()
    .rename_axis("id")
    .reset_index(name="row_count")
    .sort_values("row_count", ascending=False)
    .reset_index(drop=True)
)

top_5_share = id_counts.head(5)["row_count"].sum() / len(df) * 100
top_10_share = id_counts.head(10)["row_count"].sum() / len(df) * 100
top_10_pct_n = max(1, int(len(id_counts) * 0.10))
top_10_pct_share = id_counts.head(top_10_pct_n)["row_count"].sum() / len(df) * 100

summary = pd.DataFrame(
    {
        "metric": [
            "total_rows",
            "unique_ids",
            "mean_rows_per_id",
            "median_rows_per_id",
            "max_rows_single_id",
            "top_1_id_share_pct",
            "top_5_ids_share_pct",
            "top_10_ids_share_pct",
            "top_10pct_ids_share_pct",
        ],
        "value": [
            len(df),
            id_counts["id"].nunique(),
            round(id_counts["row_count"].mean(), 2),
            id_counts["row_count"].median(),
            id_counts["row_count"].max(),
            round(id_counts.iloc[0]["row_count"] / len(df) * 100, 2),
            round(top_5_share, 2),
            round(top_10_share, 2),
            round(top_10_pct_share, 2),
        ],
    }
)

display(summary)
display(id_counts.head(15))

top_ids = id_counts.head(15).copy()
top_ids["id"] = top_ids["id"].astype(str)

plt.figure(figsize=(12, 5))
sns.barplot(data=top_ids, x="id", y="row_count", hue="id", palette="Blues_r", legend=False)
plt.title("Top 15 IDs by row count")
plt.xlabel("id")
plt.ylabel("rows")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

cumulative = id_counts.copy()
cumulative["user_pct"] = (cumulative.index + 1) / len(cumulative) * 100
cumulative["row_share_pct"] = cumulative["row_count"].cumsum() / len(df) * 100

plt.figure(figsize=(8, 5))
plt.plot(cumulative["user_pct"], cumulative["row_share_pct"], linewidth=2)
plt.plot([0, 100], [0, 100], linestyle="--", color="gray")
plt.title("Cumulative share of rows by user rank")
plt.xlabel("Top users (%)")
plt.ylabel("Cumulative share of rows (%)")
plt.tight_layout()
plt.show()
```

```python
sema_cols = [
    "ALERT", "HAPPY", "NEUTRAL", "RESTED/RELAXED", "SAD", "TENSE/ANXIOUS", "TIRED",
    "ENTERTAINMENT", "GYM", "HOME", "HOME_OFFICE", "OTHER", "OUTDOORS", "TRANSIT", "WORK/SCHOOL",
]

sema_mask = df[sema_cols].notna().any(axis=1)
sema_df = df.loc[sema_mask].copy()

sema_counts = (
    sema_df["id"]
    .value_counts()
    .rename_axis("id")
    .reset_index(name="sema_response_count")
    .sort_values("sema_response_count", ascending=False)
    .reset_index(drop=True)
)

sema_summary = pd.DataFrame(
    {
        "metric": [
            "total_sema_rows",
            "ids_with_sema",
            "mean_sema_rows_per_id",
            "median_sema_rows_per_id",
            "max_sema_rows_single_id",
            "top_1_id_share_pct",
            "top_5_ids_share_pct",
        ],
        "value": [
            len(sema_df),
            sema_counts["id"].nunique(),
            round(sema_counts["sema_response_count"].mean(), 2),
            sema_counts["sema_response_count"].median(),
            sema_counts["sema_response_count"].max(),
            round(sema_counts.iloc[0]["sema_response_count"] / len(sema_df) * 100, 2),
            round(sema_counts.head(5)["sema_response_count"].sum() / len(sema_df) * 100, 2),
        ],
    }
)

display(sema_summary)
display(sema_counts.head(20))

plt.figure(figsize=(12, 5))
sns.barplot(
    data=sema_counts.head(15).assign(id=lambda data: data["id"].astype(str)),
    x="id",
    y="sema_response_count",
    hue="id",
    palette="crest",
    legend=False,
 )
plt.title("Top 15 IDs by SEMA survey response count")
plt.xlabel("id")
plt.ylabel("SEMA responses")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
